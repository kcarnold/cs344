```python
from fastai.vision.all import *
```

```python
root = Path("../student_data/23sp/letter-images")
dest = Path('../student_data/23sp/all-letter-images/images')
dest.mkdir(exist_ok=True, parents=True)
categories = list('abc')
for c in categories: (dest / c).mkdir(exist_ok=True)
images = get_image_files(root)
```

```python
sources = {}
for img in images:
    category = img.parent.name.lower()
    assert category in categories
    source = img.relative_to(root).parts[0]
    if source not in sources:
        sources[source] = len(sources)
    source_idx = sources[source]
    name = f'group_{source_idx}_{img.name}'
    # Copy the image data.
    (dest / category / name).write_bytes(img.read_bytes())
```
